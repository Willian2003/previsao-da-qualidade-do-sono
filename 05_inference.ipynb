```python
# Carregamento do modelo treinado para inferência
import joblib
import pandas as pd

model = joblib.load("sleep_quality_model.pkl")
```

```python
# Nomes das variáveis utilizadas no treinamento do modelo
model.feature_names_in_
```

```python
def criar_paciente(
    model,
    age,
    sleep_duration,
    activity,
    stress,
    heart_rate,
    steps,
    systolic,
    diastolic,
    gender,        # 'Male' ou 'Female'
    bmi,           # 'Normal', 'Overweight', 'Obese'
    occupation     # string igual ao one-hot
):
    # DataFrame zerado com as features do modelo
    paciente = pd.DataFrame(0, columns=model.feature_names_in_, index=[0])

    # Numéricos
    paciente.at[0, "Age"] = age
    paciente.at[0, "Sleep Duration"] = sleep_duration
    paciente.at[0, "Activity"] = activity
    paciente.at[0, "Stress Level"] = stress
    paciente.at[0, "Heart Rate"] = heart_rate
    paciente.at[0, "Steps"] = steps
    paciente.at[0, "Pressure_Systolic"] = systolic
    paciente.at[0, "Pressure_Diastolic"] = diastolic

    # Gênero
    if gender == "Male":
        paciente.at[0, "Gender_Male"] = 1

    # BMI
    if bmi == "Overweight":
        paciente.at[0, "BMI_Overweight"] = 1
    elif bmi == "Obese":
        paciente.at[0, "BMI_Obese"] = 1
    # Normal = tudo zero

    # Profissão (one-hot)
    col_prof = f"Occupation_{occupation}"
    if col_prof in paciente.columns:
        paciente.at[0, col_prof] = 1
    else:
        raise ValueError(f"Profissão inválida: {occupation}")

    return paciente

%store -r novo_paciente
```

```python
# Realização da inferência para um novo paciente
pred = model.predict(novo_paciente)
print(f"Qualidade de sono estimada: {pred[0]:.2f}")
```
