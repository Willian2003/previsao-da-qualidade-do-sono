```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Willian2003/previsao-de-disturbios-do-sono/refs/heads/main/Sleep_health_and_lifestyle_dataset.csv")
```

```python
def preprocessing(df):
    # Excluir colunas; person ID não diz nada importante; sleep disorder entrega resultado
    df.drop(columns=['Person ID', 'Sleep Disorder'], inplace=True)

    # Renomeando colunas
    df.rename(columns={
        'BMI Category': 'BMI',
        'Quality of Sleep': 'Quality',
        'Physical Activity Level': 'Activity',
        'Daily Steps': 'Steps',
    }, inplace=True)

    # Padronizando rótulos
    df['BMI'] = df['BMI'].replace({'Normal Weight': 'Normal'})

    # Separação da pressão arterial em sistólica e diastólica
    bp = df['Blood Pressure'].str.split('/', expand=True)
    df['Pressure_Systolic'] = pd.to_numeric(bp[0])
    df['Pressure_Diastolic'] = pd.to_numeric(bp[1])
    df.drop(columns=['Blood Pressure'], inplace=True)

    return df
```

```python
data = preprocessing(df)
display(data.head())
# Conjunto de dados estruturado
```

```python
data_encoded = pd.get_dummies(
    data,
    columns=["Occupation", "Gender", "BMI"],
    drop_first=True
)
# One hot encoding para evitar falsa ordenação, já que modelos como árvore não usam dados categóricos

data_encoded.to_csv("dataframe.csv", index=False)
```
